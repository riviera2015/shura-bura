```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

Аппроксимация функии потерь $\max(0, 1-x)$ с помощью сигмоиды $\frac 1 \gamma \ln (1 + \exp^{\gamma (1-x)})$:

```python
x = np.linspace(-5, 5, 100)
plt.plot(np.maximum(1-x, 0))
for gamma in np.arange(1, 7, 0.3):
    plt.plot(1/gamma*np.log(1+np.exp(gamma*(1-x))))
```
